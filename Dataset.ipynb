```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#mapa de correlacion
corr_matrix = X.corr()
plt.figure(figsize=(30, 15))
sns.heatmap(corr_matrix, annot=True, xticklabels=False, yticklabels=False)
```

```python
#diagrama torta de y
y_counts = y.value_counts()
plt.figure(figsize=(10, 10))
colors = sns.color_palette('pastel')[0:len(y_counts)] # Use a seaborn pastel color palette
plt.pie(y_counts, labels=y_counts.index, autopct='%1.1f%%', startangle=90, shadow=True, colors=colors);
```
